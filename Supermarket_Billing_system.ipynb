```python
# Supermarket Billing System

# List to store item details
items = []  

# Stores subtotal before discount
total_amount = 0  

print("Welcome to Supermarket Billing System")
name = input("Enter customer name: ")

# Loop to take multiple items as input
while True:
    # Taking item details from the user
    item_name = input("Enter item name: ")
    price = float(input("Enter price of the item: "))
    quantity = int(input("Enter quantity: "))

    # Calculating the total price for the current item
    item_total = price * quantity

    # Accumulating the total bill amount
    total_amount += item_total  

    # Storing item details in a list of dictionaries
    items.append({"name": item_name, "price": price, "quantity": quantity, "total": item_total})

    # Asking the user if they want to add more items
    repeat = input("Do you want to add more items? (yes/no) ")
    if repeat.lower() == "no":  
        break  # Exit the loop if the user says "no"

# Applying Discount based on total bill amount
discount = 0

if total_amount > 500:
    discount = total_amount * 0.10  # 10% discount for bills over ₹500
elif total_amount > 200:
    discount = total_amount * 0.05  # 5% discount for bills over ₹200

# Final amount after applying discount
final_amount = total_amount - discount  

# Printing the Bill
print("\n===== Supermarket Bill =====")
print(f"Customer: {name}")

# Printing Itemized Bill Header
print("\nItemized Bill:")
print("{:<15} {:<10} {:<10} {:<10}".format("Item", "Price", "Quantity", "Total"))
print("-" * 50)

# Iterating through each item and printing its details
for item in items:
    print("{:<15} {:<10} {:<10} {:<10}".format(item["name"], item["price"], item["quantity"], item["total"]))

print("-" * 50)
print(f"Subtotal: {total_amount:.2f}")  
print(f"Discount Applied: {discount:.2f}")  
print(f"Final Amount to Pay: {final_amount:.2f}")  
print("=============================")
print("Thank you for shopping with us!")

# Sample Input and Output for Demonstration
print("\n===== SAMPLE INPUT/OUTPUT =====")

sample_input = """\
Enter customer name: Alex
Enter item name: Apple
Enter price of the item: 30
Enter quantity: 2
Do you want to add more items? (yes/no) yes
Enter item name: Bread
Enter price of the item: 50
Enter quantity: 1
Do you want to add more items? (yes/no) yes
Enter item name: Rice
Enter price of the item: 250
Enter quantity: 1
Do you want to add more items? (yes/no) no
"""

sample_output = """\
===== Supermarket Bill =====
Customer: Alex

Itemized Bill:
Item            Price      Quantity   Total     
--------------------------------------------------
Apple           30.0       2          60.0      
Bread           50.0       1          50.0      
Rice            250.0      1          250.0     
--------------------------------------------------
Subtotal: 360.00
Discount Applied: 18.00
Final Amount to Pay: 342.00
=============================
Thank you for shopping with us!
"""

print("\n--- Sample Input ---")
print(sample_input)

print("\n--- Sample Output ---")
print(sample_output)
```
